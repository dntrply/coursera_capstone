```
#check to see if BeautifulSoup is installed properly
from bs4 import BeautifulSoup
```

```
# Read in the URL
import requests
```

```
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```
#render the data in BeautiulSoup
soup = BeautifulSoup(source, 'lxml')

#print - to make sure we have the data
print(soup.prettify())
```

```
# Access the table data -print the same to enusre we see the content
table = soup.tbody
print(table.prettify())
```

```
# The first row has the header information
table_header_row = table.find('tr')
```

```
table_header_row
```

```
# Get to the column names
column_names = [table_cell.text.rstrip() for table_cell in table.find('tr').find_all('th')]
```

```
column_names
```

```
# Get all rows
table_rows = table.find_all('tr')
```

```
table_rows[0:7]
```

```
# the second row onwards has the data in the columns data
# note that since we are getting the 'td' text, we do not have to worry about <a> </a> attribute
toronto_df = pd.DataFrame(([table_cell.text.rstrip() for table_cell in table_row.find_all('td')] for table_row in table_rows[1:]), columns=column_names)
```

```
toronto_df.describe()
```

```
toronto_df.head()
```

```
# we need to remove all entries wehre the Borough has the value Not Assigned.
# This can be accomplished a couple of ways
# In the forst way - commented out - a mask is created
toronto_df['Borough'] != 'Not assigned'
```

```
# ths mask can then be applied to the dataframe and a new dataframe created with the Not assigned entries removed
toronto_df_borough = toronto_df[toronto_df['Borough']!='Not assigned']
```

```
# Alternatively, the second approach is to:
# replace all the 'Not assigned' with a NAN (across all columns)
# drop the NAN along the appropriate columns (Borough in this case)
# we now have  anew dataframe with the Not assigned entries from the Borough column removed
toronto_df_new_borough = toronto_df.replace('Not assigned', np.nan).dropna(subset=['Borough'])
```

```
# We will focus on the toronto_df_new_borough from now on
# Let's first find all the 'uniqie' neighbourhoods - this is just exploring the data
toronto_df_new_borough['Neighbourhood'].unique()
```

```
toronto_df_new_borough.head(10)
```

```
# the next thing to do is to replace the Not Assigned on the Neighbourhood (note that it is now a NAN) withthe value in the Borough
# We will first start with a mask of all NAN on teh Neighbourhood column
```

```
subset = toronto_df_new_borough['Neighbourhood'].isna()
```

```
subset
```

```
subset.head(10)
```

```
# The values we need to replace with are from the corresponding Borough column
replace_values = toronto_df_new_borough[subset]['Borough']
```

```
replace_values
```

```
# We will next use loc and pass it the subset so as to replace the appropriate values
toronto_df_new_borough.loc[subset, 'Neighbourhood'] = replace_values
```

```
toronto_df_new_borough.head(10)
```

```
# The next task is to have a comma separated neighborhood values for the same postal code area
# This is accomplished by using the groupby method. Note that we are doing two levels of grouping
# the first by Postal code and next by Borough
```

```
grouped = toronto_df_new_borough.groupby(by = ['Postcode', 'Borough'])
```

```
# Set an object so as to make it easier to extract the information
reduced_series = grouped['Neighbourhood'].apply(lambda NHoods: ', '.join(NHoods))
```

```
lat_list = []
long_list = []

for postcode in toronto_df_reduced['Postcode']:

  request = 'http://dev.virtualearth.net/REST/v1/Locations/CA/-/' + postcode + '/-/-?&maxResults=5&key=' + api_key
  response = requests.get(request)
  json_response = json.loads(response.text)
  
  # get the latitude and longitude
  lat = json_response['resourceSets'][0]['resources'][0]['point']['coordinates'][0]
  long = json_response['resourceSets'][0]['resources'][0]['point']['coordinates'][1]

  lat_list.append(lat)
  long_list.append(long)
```
